## Reading files
Assuming you have a file containing both data and geometry (e.g. GeoPackage, GeoJSON, Shapefile), you can read it using `geopandas.read_file()`, which automatically detects the filetype and creates a `GeoDataFrame`. This tutorial uses the `"nybb"` dataset, a map of New York boroughs, which is available through the `geodatasets` package. Therefore, we use `geodatasets.get_path()` to download the dataset and retrieve the path to the local copy.

```python
import geopandas
import warnings
warnings.filterwarnings("ignore")

url = "data/GeoMapData_CN/province/210000.json"
df = geopandas.read_file(url)

df
```

```python
df.explore()
```

```python
# https://www.runoob.com/matplotlib/matplotlib-zh.html
import matplotlib.pyplot as plt
import matplotlib
zhfont = matplotlib.font_manager.FontProperties(fname="../font/SourceHanSansSC-Medium.otf") 
df['area'] = df.area
df["centroid"] = df.centroid
ax = df.plot('area', legend=False, figsize=(12, 9))
df["centroid"].plot(ax=ax, color="white")
texts = df.apply(lambda r: {'s': r['name'], 'x': r.centroid.x + 0.1, 'y': r.centroid.y - 0.02}, axis=1)
for text in texts:
    ax.text(**text, fontproperties=zhfont, color='white')
```
